```python
import pickle
import pandas as pd
import numpy as np
import glob
import faiss
import time
```

```python
with open("retrievals/adav2_embeddings_all.pkl", "rb") as input_file:
    adav2_corpus = pickle.load(input_file)
```

```python
with open("retrievals/adav2_query_embeddings.pkl", "rb") as input_file:
    adav2_queries = pickle.load(input_file)
```

```python
adav2_corpus = adav2_corpus.astype(np.float32)
adav2_queries = adav2_queries.astype(np.float32)
```

### Create the FAISS index

```python
corpus_index = faiss.IndexFlatIP(adav2_corpus.shape[-1])
```

```python
type(adav2_corpus)
```

```python
start_time = time.time()

corpus_index.add(adav2_corpus)

print("FAISS index creation elapsed time: {}".format(time.time() - start_time))
```

## Search the queries in the index

```python
search_results = corpus_index.search(adav2_queries, 1000)
```
