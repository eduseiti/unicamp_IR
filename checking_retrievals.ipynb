#### Now compare the top_k documents returned by mT5 and Colbertx

```python
top_k=10

results = []

for group_name, group_df in mT5_df.groupby('query_id', sort=False):
    
    print("Handling query {}...".format(group_name))
    
    mT5_docs = group_df.iloc[:top_k]['doc_id'].to_numpy()
    
    if group_name >= 100:
        colbertx_docs = retrievals_100_199_df[retrievals_100_199_df['query_id'] == group_name].sort_values(['score_znorm'], ascending=[False]).iloc[:top_k]['clueweb_id'].to_numpy()
    else:
        colbertx_docs = retrievals_000_099_df[retrievals_000_099_df['query_id'] == group_name].sort_values(['score_znorm'], ascending=[False]).iloc[:top_k]['clueweb_id'].to_numpy()
        
    common_docs = np.intersect1d(mT5_docs, colbertx_docs)
        
    results.append({'query_id': group_name,
                    'mT5_docs': np.sort(mT5_docs),
                    'colbertx_docs': np.sort(colbertx_docs),
                    'mt5_colbertx_common_docs': np.sort(common_docs),
                    'mt5_colbertx_common_proportion': common_docs.shape[0] / top_k})
```

```python
results_df = pd.DataFrame(results)
```

#### Now compare the top_k documents returned by mT5, Colbertx and SPLADE

```python
top_k=10

results_splade = []

i = 0

for group_name, group_df in mT5_df.groupby('query_id', sort=False):
    
    print("Handling query {}...".format(group_name))
    
    mT5_docs = group_df.iloc[:top_k]['doc_id'].to_numpy()
    
    splade_docs = splade_df[splade_df['query_id'] == group_name].sort_values(['score'], ascending=[False]).iloc[:top_k]['clueweb_id'].to_numpy()
        
    mt5_splade_common_docs = np.intersect1d(mT5_docs, splade_docs)
    colbertx_splade_common_docs = np.intersect1d(results[i]['colbertx_docs'], splade_docs)
    
    common_docs = np.intersect1d(mt5_splade_common_docs, colbertx_splade_common_docs)
    
    results_splade.append({'query_id': group_name,
                           'mT5_docs': np.sort(mT5_docs),
                           'colbertx_docs': np.sort(splade_docs),
                           'mt5_splade_common_docs': np.sort(mt5_splade_common_docs),
                           'mt5_splade_common_proportion': mt5_splade_common_docs.shape[0] / top_k,
                           'colbertx_splade_common_docs': np.sort(colbertx_splade_common_docs),
                           'colbertx_splade_common_proportion': colbertx_splade_common_docs.shape[0] / top_k,
                           'common_docs': np.sort(common_docs),
                           'common_proportion': common_docs.shape[0] / top_k
                          })
    i += 1
```

```python
results_splade_df = pd.DataFrame(results_splade)
```
