### ColbertX retrievals

```python
retrievals_100_199_df = pd.read_csv(os.path.join(RETRIEVALS_FOLDER, "colbertx_all_znormalized_validation_scores_100_199.tsv"), sep='\t')
```

```python
np.sort(retrievals_100_199_df['query_id'].unique())
```

```python
retrievals_100_199_df['query_id'].unique().shape
```

```python
top_k=10

results = []

for group_name, group_df in mT5_df.groupby('query_id', sort=False):
    
    print("Handling query {}...".format(group_name))
    
    mT5_docs = group_df.iloc[:top_k]['doc_id'].to_numpy()
    
    if group_name < 100:
        colbertx_docs = retrievals_100_199_df[retrievals_100_199_df['query_id'] == group_name + 100].sort_values(['score_znorm'], ascending=[False]).iloc[:top_k]['clueweb_id'].to_numpy()
    else:
        colbertx_docs = retrievals_000_099_df[retrievals_000_099_df['query_id'] == group_name - 99].sort_values(['score_znorm'], ascending=[False]).iloc[:top_k]['clueweb_id'].to_numpy()
        
    common_docs = np.intersect1d(mT5_docs, colbertx_docs)
        
    results.append({'query_id': group_name,
                    'mT5_docs': np.sort(mT5_docs),
                    'colbertx_docs': np.sort(colbertx_docs),
                    'common_docs': np.sort(common_docs),
                    'common_proportion': common_docs.shape[0] / top_k})
```

```python
results_df = pd.DataFrame(results)
```
