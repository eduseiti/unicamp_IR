## Check agreement per question

```python
a1_df = pd.read_csv(os.path.join(ANNOTATIONS_FOLDER, "admin_240_annotations_with_questions.tsv"), sep='\t')
a2_df = pd.read_csv(os.path.join(ANNOTATIONS_FOLDER, "Eduardo_240_annotations_with_questions.tsv"), sep='\t')
a3_df = pd.read_csv(os.path.join(ANNOTATIONS_FOLDER, "Leodecio_240_annotations_with_questions.tsv"), sep='\t')
```

```python
def check_agreement_per_questions(evaluation_a, evaluation_b, prefix=None, correlation_function="cohen_kappa_score"):
    merged_df = evaluation_a.merge(evaluation_b, left_on='doccano_id', right_on='doccano_id')[['query_x', 'passage_x', 'passage_id_x', 'score_x', 'score_y']]
    
    correlations = []
    
    for group_name, group_df in merged_df.groupby('query_x'):

        correlation_result = np.array(globals()[correlation_function](group_df['score_x'], group_df['score_y']))

        if correlation_result.shape[0] > 1:
            correlations.append({'query': group_name,
                                 '{}{}'.format(correlation_function, prefix): correlation_result[0],
                                 '{}{}_pvalue'.format(correlation_function, prefix): correlation_result[1]})
        else:
            correlations.append({'query': group_name,
                                '{}{}'.format(correlation_function, prefix): correlation_result})
        
    return merged_df, pd.DataFrame(correlations)
```

```python
res_a1_a2 = check_agreement_per_questions(a1_df, a2_df, "_a1_a2", correlation_function="spearmanr")
res_a1_a3 = check_agreement_per_questions(a1_df, a3_df, "_a1_a3", correlation_function="spearmanr")
res_a2_a3 = check_agreement_per_questions(a2_df, a3_df, "_a2_a3", correlation_function="spearmanr")
```

```python
def consolidate_correlataions(correlations_dfs_list):
    combined_res_df = pd.concat([correlations_dfs_list[0].iloc[:, :2]] + [df.iloc[:, -2] for df in correlations_dfs_list[1:]], axis=1)
    
    dfs_mean = combined_res_df.iloc[:, 1:].mean(axis=1)
    dfs_std = combined_res_df.iloc[:, 1:].std(axis=1)
    dfs_median = combined_res_df.iloc[:, 1:].median(axis=1)
    
    combined_res_df['mean'] = dfs_mean
    combined_res_df['std'] = dfs_std
    combined_res_df['median'] = dfs_median
    
    return combined_res_df
```

```python
combined_res_df = consolidate_correlataions([res_a1_a2[1], res_a1_a3[1], res_a2_a3[1], res_gpt4_2_a1[1], res_gpt4_2_a2[1], res_gpt4_2_a3[1]])
```
