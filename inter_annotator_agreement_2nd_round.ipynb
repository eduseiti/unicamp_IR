```python
import pandas as pd
import numpy as np
import glob

import os

from sklearn.metrics import cohen_kappa_score

from scipy.stats import spearmanr
```

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', 200)
```

```python
ANNOTATIONS_FOLDER="anotações_humanas"
```

```python
annotations = glob.glob(os.path.join(ANNOTATIONS_FOLDER, "*.csv"))
```

```python
annotations
```

```python
a1_df = pd.read_csv(annotations[4])
```

```python
a1_df['label'].unique()
```

```python
classes_map = {
    'Perfect': 3,
    'Highly Relevant': 2,
    'Relevant': 1,
    'Irrelevant': 0
}
```

```python
a1_df['score'] = a1_df['label'].map(classes_map)
```

```python
a1_df
```

```python
a2_df = pd.read_csv(annotations[5])
a3_df = pd.read_csv(annotations[0])
```

```python
a2_df['score'] = a2_df['label'].map(classes_map)
a3_df['score'] = a3_df['label'].map(classes_map)
```

## Check annotators agreement

```python
cohen_kappa_score(a1_df['score'], a2_df['score'])
```

```python
cohen_kappa_score(a1_df['score'], a3_df['score'])
```

```python
cohen_kappa_score(a2_df['score'], a3_df['score'])
```

## Check annotators correlation against LLM evaluations

### Start with GPT3.5 evaluations

```python
gpt3_df = pd.read_csv(os.path.join(ANNOTATIONS_FOLDER, "gpt_3.5_evaluations_120_samples.tsv"), sep='\t')
```

```python
gpt3_df['score'].to_numpy()
```

```python
map_gpt_score={
    10: 3,
     9: 3,
     8: 2,
     7: 2,
     6: 2,
     5: 2,
     4: 2,
     3: 1,
     2: 1,
     1: 1,
     0: 0,
}
```

```python
gpt3_df['redux_score'] = gpt3_df['score'].map(map_gpt_score)
```

```python
gpt3_df['redux_score'].to_numpy()
```

```python
print(cohen_kappa_score(a1_df['score'], gpt3_df['redux_score'].to_numpy()))

print(cohen_kappa_score(a2_df['score'], gpt3_df['redux_score'].to_numpy()))
print(cohen_kappa_score(a3_df['score'], gpt3_df['redux_score'].to_numpy()))
```

```python
gpt3_redux_df = pd.read_csv(os.path.join(ANNOTATIONS_FOLDER, "gpt_3_evaluations_120_samples_2nd_round.tsv"), sep='\t')
```

```python
print(cohen_kappa_score(a1_df['score'], gpt3_redux_df['score'].to_numpy()))
print(cohen_kappa_score(a2_df['score'], gpt3_redux_df['score'].to_numpy()))
print(cohen_kappa_score(a3_df['score'], gpt3_redux_df['score'].to_numpy()))
```

### Now the GPT4 evaluations

```python
gpt4_df = pd.read_csv(os.path.join(ANNOTATIONS_FOLDER, "gpt_4_evaluations_120_samples.tsv"), sep='\t')
```

```python
gpt4_redux_df = pd.read_csv(os.path.join(ANNOTATIONS_FOLDER, "gpt_4_evaluations_120_samples_2nd_round.tsv"), sep='\t')
```

```python
print(cohen_kappa_score(a1_df['score'], gpt4_redux_df['score'].to_numpy()))
print(cohen_kappa_score(a2_df['score'], gpt4_redux_df['score'].to_numpy()))
print(cohen_kappa_score(a3_df['score'], gpt4_redux_df['score'].to_numpy()))
```
