### Merge both results

```python
results = []

for group_name, group_df in first_df.groupby('query_id'):

    print("\nHandling query_id={}".format(group_name))
    
    query_df = second_df[second_df['query_id'] == group_name]
    query_docs = []
    passages = group_df['passage_id'].to_numpy()
    e5_passages = query_df['passage_id'].to_numpy()
  
    print("-- Existing passages={}\n-- BM25 only passages={}".format(passages.shape[0],
                                                                     10 - np.intersect1d(passages, e5_passages).shape[0]))
    
    for i, row in group_df.iterrows():

        # print(row)
        
        query_docs.append({"query_id": group_name,
                           "query": row['query'],
                           "passage_id": row['passage_id'],
                           "mT5": row['mT5'],
                           "splade": row['splade'],
                           "colbertx": row['colbertx'],
                           "bm25_e5": row['bm25_e5'],
                           "bm25": row['bm25'],
                           "e5": row['e5'],
                           "e5_base": row['passage_id'] in e5_passages,
                           "passage": row['passage'],
                           "score": row['score'],
                           "reasoning": row['reasoning'],
                           "usage": row['usage'],
                           "cost": row['cost'],
                           "duration": row['duration'],
                           "saved_cost": row['saved_cost']})

    for i, row in query_df.iterrows():
        if row['passage_id'] not in passages:
            query_docs.append({"query_id": group_name,
                               "query": row['query'],
                               "passage_id": row['passage_id'],
                               "mT5": False,
                               "splade": False,
                               "colbertx": False,
                               "bm25_e5": False,
                               "bm25": False,
                               "e5": False,
                               "e5_base": True,
                               "passage": row['passage'],
                               "score": row['score.1'],
                               "reasoning": row['reasoning'],
                               "usage": row['usage'],
                               "cost": row['cost'],
                               "duration": row['duration'],
                               "saved_cost": row['saved_cost']})                               

    print("-- Final number of passages={}".format(len(query_docs)))

    results += query_docs
```

```python
results_df = pd.DataFrame(results)
```
