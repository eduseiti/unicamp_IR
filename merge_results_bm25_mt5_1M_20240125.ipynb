### Merge both results

```python
results = []

for group_name, group_df in first_df.groupby('query_id'):

    print("\nHandling query_id={}".format(group_name))

    query_df = second_df[second_df['query_id'] == group_name]
    query_docs = []
    passages = group_df['passage_id'].to_numpy()
    bm25_mt5_passages = query_df['passage_id'].to_numpy()

    print("-- Existing passages={}\n-- BM25 mT5 1M only passages={}".format(passages.shape[0],
                                                                            10 - np.intersect1d(passages, bm25_mt5_passages).shape[0]))

    for i, row in group_df.iterrows():

        # print(row)

        query_docs.append({"query_id": group_name,
                           "query": row['query'],
                           "passage_id": row['passage_id'],
                           "mT5": row['mT5'],
                           "splade": row['splade'],
                           "colbertx": row['colbertx'],
                           "bm25_e5": row['bm25_e5'],
                           "bm25": row['bm25'],
                           "e5": row['e5'],
                           "e5_base": row['e5_base'],
                           "splade_mT5": row['splade_mT5'],
                           "bm25_1M": row['bm25_1M'],
                           "bm25_1M_massiveweb": row['bm25_1M_massiveweb'],
                           "colbertx_1M_massiveweb": row['colbertx_1M_massiveweb'],
                           "e5_colbertx_rrf": row['e5_colbertx_rrf'],
                           "e5_splade_rrf": row['e5_splade_rrf'],
                           "e5_splade_rrf_fix": row['e5_splade_rrf_fix'],
                           "adav2": row['adav2'],
                           "bm25_mT5_1M_massiveweb": row['passage_id'] in bm25_mt5_passages,
                           "passage": row['passage'],
                           "score": row['score'],
                           "reasoning": row['reasoning'],
                           "usage": row['usage'],
                           "cost": row['cost'],
                           "duration": row['duration'],
                           "saved_cost": row['saved_cost']})

    for i, row in query_df.iterrows():
        if row['passage_id'] not in passages:
            query_docs.append({"query_id": group_name,
                               "query": row['query'],
                               "passage_id": row['passage_id'],
                               "mT5": False,
                               "splade": False,
                               "colbertx": False,
                               "bm25_e5": False,
                               "bm25": False,
                               "e5": False,
                               "e5_base": False,
                               "splade_mT5": False,
                               "bm25_1M": False,
                               "bm25_1M_massiveweb": False,
                               "colbertx_1M_massiveweb": False,
                               "e5_colbertx_rrf": False,
                               "e5_splade_rrf": False,
                               "e5_splade_rrf_fix": False,
                               "adav2": False,
                               "bm25_mT5_1M_massiveweb": True,
                               "passage": row['passage'],
                               "score": row['score.1'],
                               "reasoning": row['reasoning'],
                               "usage": row['usage'],
                               "cost": row['cost'],
                               "duration": row['duration'],
                               "saved_cost": row['saved_cost']})

    print("-- Final number of passages={}".format(len(query_docs)))

    results += query_docs
```

```python
def check_all_returned_passages_score_distribution(results_df, retrievers_columns):

    system_stats = []

    for retriever in retrievers_columns:
        single_system_df = results_df[results_df[retriever]]

        score_distribution = np.unique(single_system_df['score'].to_numpy(), return_counts=True)

        system_stats.append({"retriever": retriever,
                             "total": score_distribution[1].sum(),
                             "0": score_distribution[1][0],
                             "0 (%)": score_distribution[1][0]/score_distribution[1].sum(),
                             "1": score_distribution[1][1],
                             "1 (%)": score_distribution[1][1]/score_distribution[1].sum(),
                             "2": score_distribution[1][2],
                             "2 (%)": score_distribution[1][2]/score_distribution[1].sum(),
                             "3": score_distribution[1][3],
                             "3 (%)": score_distribution[1][3]/score_distribution[1].sum()})

    return pd.DataFrame(system_stats)
```

```python
def check_returned_passages_score_distribution(results_df, retrievers_columns):

    single_system_df = results_df.loc[results_df[retrievers_columns].sum(axis=1) == 1]

    single_system_stats = []

    for retriever in retrievers_columns:
        score_distribution = np.unique(single_system_df[single_system_df[retriever]]['score'].to_numpy(), return_counts=True)

        single_system_stats.append({"retriever": retriever,
                                    "total": score_distribution[1].sum(),
                                    "0": score_distribution[1][0],
                                    "0 (%)": score_distribution[1][0]/score_distribution[1].sum(),
                                    "1": score_distribution[1][1],
                                    "1 (%)": score_distribution[1][1]/score_distribution[1].sum(),
                                    "2": score_distribution[1][2],
                                    "2 (%)": score_distribution[1][2]/score_distribution[1].sum(),
                                    "3": score_distribution[1][3],
                                    "3 (%)": score_distribution[1][3]/score_distribution[1].sum()})

    return pd.DataFrame(single_system_stats)
```

```python
single_distribution_df = check_returned_passages_score_distribution(results_df, ['mT5', 'splade', 'colbertx', 'bm25_e5', 'bm25', 'e5', 'e5_base', 'splade_mT5', 'bm25_1M', 'bm25_1M_massiveweb', 'colbertx_1M_massiveweb', 'e5_colbertx_rrf', 'e5_splade_rrf', 'e5_splade_rrf_fix', 'adav2', 'bm25_mT5_1M_massiveweb'])
```
