```python
ALL_VALIDATION_QUERIES="all_validation_queries.tsv"
```

```python
BM25_RETRIEVAL="200-queries-passagens-by-BM25-pt_v2_with_query_id.tsv"
BM25_MT5_RETRIEVAL="200-queries-passagens-by-BM25-mT5_v4_with_query_id.tsv"
BM25_E5_RETRIEVAL="bm25_e5_docs_000_999_50_sampled_queries.tsv"
SPLADE_RETRIEVAL="splade_pt_result_with_clueweb_id.tsv"
COLBERTX_RETRIEVAL="colbertx_all_znormalized_validation_scores_000_199.tsv"
E5_RETRIEVAL="e5_docs_000_999_50_sampled_queries_score_inv.tsv"
E5_BASE_RETRIEVAL="e5_base_docs_000_999_50_sampled_queries.tsv"
```

```python
all_retrievals=[
    {"retriever": "BM25", "file": BM25_RETRIEVAL, "passage_field": "passagem-id", "score_field": None},
    {"retriever": "BM25+MT5", "file": BM25_MT5_RETRIEVAL, "passage_field": "doc_id", "score_field": None},
    {"retriever": "BM25+E5", "file": BM25_E5_RETRIEVAL, "passage_field": "passage_id", "score_field": "score"},
    {"retriever": "SPLADE", "file": SPLADE_RETRIEVAL, "passage_field": "clueweb_id", "score_field": "score"},
    {"retriever": "COLBERTX", "file": COLBERTX_RETRIEVAL, "passage_field": "clueweb_id", "score_field": "score_znorm"},
    {"retriever": "E5", "file": E5_RETRIEVAL, "passage_field": "passage_id", "score_field": "score_inv"},
    {"retriever": "E5 BASE", "file": E5_BASE_RETRIEVAL, "passage_field": "passage_id", "score_field": "score_inv"}
]

ALL_RETRIEVERS = np.array([results['retriever'] for results in all_retrievals])
```

```python
SAMPLED_QUERIES=[  1,   2,   9,  11,  13,  15,  17,  20,  21,  22,  24,  26,  28,
                  47,  49,  51,  54,  60,  62,  64,  68,  84,  95,  98, 105, 113,
                 115, 117, 126, 127, 128, 136, 147, 152, 153, 154, 160, 161, 163,
                 167, 170, 172, 180, 181, 182, 189, 193, 195, 196, 199]
```

```python
for results in all_retrievals:
    print("Reading results from {}...".format(results['retriever']))
    
    results_df = pd.read_csv(os.path.join(RETRIEVALS_FOLDER, results['file']), sep='\t')
    
    display(results_df.head())
    
    results['df'] = results_df
```

```python
all_queries_df = pd.read_csv(os.path.join(RETRIEVALS_FOLDER, ALL_VALIDATION_QUERIES), sep="\t")
```

```python
start_pos = 0
how_many = 10
```

```python
compiled_queries = []

for query_id in SAMPLED_QUERIES:
    print("\nHandling query {}...".format(query_id))
    
    query_text = all_queries_df[all_queries_df['query_id'] == query_id]['query'].to_numpy()[0]
    
    query_docs = {}
    
    for i, results in enumerate(all_retrievals):
        print("\n-- Checking {}...".format(results['retriever']))
        
        all_other_retrievals = np.delete(ALL_RETRIEVERS, i)
        
        # print(all_other_retrievals)
        
        results_df = results['df']
        
        if results['score_field'] is None:
            results_docs = results_df[results_df['query_id'] == query_id].iloc[start_pos:(start_pos + how_many)][results['passage_field']].to_numpy()            
        else:
            results_docs = results_df[results_df['query_id'] == query_id].sort_values([results['score_field']], 
                                                                                      ascending=[False]).iloc[start_pos:(start_pos + how_many)][results['passage_field']].to_numpy()
    
        print("results_docs.shape={}".format(results_docs.shape))
    
        for doc in results_docs:
            if doc in query_docs:

                print(">>>> document {} already processed...".format(doc))

                # print(query_docs[doc].keys())
                
                query_docs[doc][results['retriever']] = True
            else:
                query_docs[doc] = {"query_id": query_id,
                                   "query": query_text,
                                   "passage_id": doc
                                  }
                
                query_docs[doc][results['retriever']] = True
                
                for other_retriever in all_other_retrievals:
                    query_docs[doc][other_retriever] = False
                    
    compiled_queries += list(query_docs.values())
```
